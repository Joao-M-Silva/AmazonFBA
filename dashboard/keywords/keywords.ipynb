```python
df["Magnet IQ Score"] = df["Magnet IQ Score"].apply(lambda x: x.replace(",", "")).astype(int).apply(lambda x: -x)
df["Search Volume"] = df["Search Volume"].apply(lambda x: x.replace(",", "")).astype(int)
df["Competing Products"] = df["Competing Products"].apply(lambda x: x.replace(",", "").replace(">", "")).astype(int)
```

```python
df_sorted = df.sort_values(["Competing Products", "Magnet IQ Score"])
```

```python
df_sorted.head()
```
